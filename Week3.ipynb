```python
import pandas as pd

# Write your code below
df = pd.read_csv('data.csv')
numeric_columns = df.columns
# to find the 25th percentile and 75th percentile for the numerical columns.
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
iqr = Q3 - Q1

 #Finding lower and upper bounds for all values. All values outside these bounds are outliers
lower_bound = Q1 - 1.5*iqr
upper_bound = Q3 + 1.5*iqr

# 11 Percentage of outliers in each column
outlier_columns = list(df.columns[(((df < lower_bound) | (df > upper_bound)).sum() / df.shape[0] > 0.11)])
```

```python
import pandas as pd
import numpy as np

# Write your code below
df = pd.read_csv('data.csv')

outlier_columns = ['Ad Spend', 'Profit Margin', 'Revenue']

# to find the 25th percentile and 75th percentile for the numerical columns.
Q1 = df[outlier_columns].quantile(0.25)
Q3 = df[outlier_columns].quantile(0.75)
iqr = Q3 - Q1

 #Finding lower and upper bounds for all values. All values outside these bounds are outliers
lower_bound = Q1 - 1.5*iqr
upper_bound = Q3 + 1.5*iqr

for col in outlier_columns:
    df[col] = np.clip(df[col], lower_bound[col], upper_bound[col])
```

```python
import pandas as pd

# Write your code below
df = pd.read_csv('data.csv')
min_dist = df['Distance'].min()
max_dist = df['Distance'].max()

def normalize(X):
    return ((X-min_dist)/(max_dist-min_dist))
df['Distance'] = df['Distance'].apply(normalize)
```

```python
import pandas as pd

# Write your code below
df = pd.read_csv('data.csv')
sq_price = df['Price']**2
sum_sq_price = sq_price.sum()
norm = sum_sq_price**0.5

def normalize_price(X):
    return (X / norm)
df['Price'] = df['Price'].apply(normalize_price)
```
